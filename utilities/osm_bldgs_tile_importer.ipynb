```python
# get all the OSM buildings in a country's city
def get_country_bldgs_df(country_code):
    res_df = pd.DataFrame(columns=['id', 'type', 'properties', 'geometry'])
    for tile in tiles[country_code]:
        osmb_url = "https://data.osmbuildings.org/0.2/anonymous/tile/15/{}/{}.json".format(tile[0], tile[1])
        response = requests.get(osmb_url)
        be = response.json()
        res_df = res_df.append(pd.DataFrame(be["features"]))
    res_df.loc[:, 'country'] = country_code
    return res_df
```

```python
# Berlin, de
de = get_country_bldgs_df('de')
```

```python
# London, gb
gb = get_country_bldgs_df('gb')
```

```python
# Amsterdam, nl
nl = get_country_bldgs_df('nl')
```

```python
# Dublin, ie
ie = get_country_bldgs_df('ie')
```

```python
# Paris, fr
fr = get_country_bldgs_df('fr')
```

```python
df = de.append(gb).append(nl).append(ie).append(fr)
```

```python
df.reset_index(inplace=True, drop=True)
df.drop('type', axis=1, inplace=True)

df.drop_duplicates('id', inplace=True)
df.columns = ["id", "properties", "geo", "country"]
```
